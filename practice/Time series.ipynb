#### Real-world examples of time series data

#### Explain key characteristics: Ordered, Dependent, and Time-dependent.

*1.Ordered*: In time series data, the sequence of observations matters a lot. The order in which the data points were collected is crucial information. This order is what distinguishes time series data from other types of data. Analyzing data out of order can lead to completely wrong conclusions.
- Imagine you're tracking the daily temperature in your city. If you mix up the order of the readings, you'll lose the ability to see temperature trends, like a heatwave or a cold snap.

*2.Dependent*: Data points in a time series are often related to each other. Past values can influence current and future values. This dependence is a key feature that statistical models for time series need to account for. Ignoring it can lead to inaccurate forecasts.
The price of a stock today is likely influenced by its price yesterday, and the day before that, and so on. This is why stock prices can form trends and patterns over time.

*3. Time-dependent:* The statistical properties of the data (like the average or variability) can change over time. This is a challenge that makes analyzing time series data more complex. You can't assume that the patterns you see in one period will necessarily hold true for another. The demand for ice cream might be high in the summer and low in the winter. This means the average demand is time-dependent.

# Trend in Time Series

In time series analysis
- In time series analysis, trend refers to the long-term movement or direction of the data over time. It represents a sustained increase or decrease in the values of the series, which may be due to factors like economic growth, market demand, or other structural changes. A trend can either be increasing, decreasing, or constant.

Types of Trend

Linear Trend: 
- A linear trend indicates a constant rate of change in the data over time. The relationship between time and the variable is represented by a straight line.

Exponential Trend:
- An exponential trend is when the data grows or decays at a constant percentage rate over time. This results in a curve rather than a straight line, as the rate of change accelerates or decelerate

Quadratic Trend:
- A quadratic trend involves a parabolic (U-shaped or inverted U-shaped) curve, where the rate of change itself changes over time. This means the trend is not constant but accelerates or decelerates in a nonlinear manner.

# Seasonality in Time Series

Seasonality in Time Series
- Seasonality refers to regular, predictable patterns or fluctuations that occur in a time series data at consistent intervals, typically due to factors such as the calendar, weather, or other cyclical phenomena. It represents the repeating short-term fluctuations in a dataset that are expected at certain times of the year, month, week, or even day.

Importance of Seasonality
- Forecasting: Recognizing and understanding seasonality helps to make more accurate predictions for future periods. 
- Business Planning: Companies can adjust their operations (e.g., inventory, staffing) to match the expected seasonal fluctuations.
- Demand Forecasting: Seasonal trends help in predicting high-demand and low-demand periods, allowing companies to optimize their supply chain and resources.
- Marketing and Promotions: Identifying seasonal patterns aids in timing marketing campaigns, promotional offers, and product launches.

Types of Seasonality

- Additive Seasonality:
Additive seasonality occurs when the seasonal effect is constant over time. The magnitude of the seasonal fluctuations does not change with the level of the trend.
In this case, the seasonal variations are added to the trend.

- Multiplicative Seasonality:
Multiplicative seasonality occurs when the seasonal effect changes proportionally with the level of the trend. The magnitude of the seasonal variations grows or shrinks with the level of the trend (i.e., the seasonal effect increases when the overall trend is higher).
In this case, the seasonal variations are multiplied by the trend.

Noise in Time Series Data
- Noise refers to the random fluctuations or irrelevant variations in the data that do not have any predictable pattern, trend, or seasonal component. In time series data, noise is typically the portion of the data that cannot be explained by the trend, seasonality, or other deterministic factors.

Measurement Errors:
- Instrumental errors or data recording issues can introduce noise. For example, if there is a malfunction in a sensor or if the measurement tool is inaccurate, it can lead to random 

External Environmental Factors:
- Unpredictable weather (e.g., sudden storms or temperature extremes) can cause sudden fluctuations in industries like agriculture, transportation, and energy.

Random Events:
- Random incidents or outliers like unexpected accidents, product recalls, or sudden changes in consumer behavior can cause noise. These events are often not part of a broader trend or seasonal pattern and cannot be predicted.
For example, a one-off product failure in a factory 

Economic Shocks:
- Sudden economic shifts or financial crises, such as a market crash, an economic recession, or an unexpected rise in commodity prices, can introduce noise. These events cause abrupt changes that might not be explained by normal trends or seasonal patterns.
Market Volatility:

In financial markets (stocks, commodities, etc.), 
- market volatility can introduce noise, especially when there's high trading volume, speculation, or reactions to news that do not follow a clear pattern.
Sampling Variability:




Dealing with Noise in Time Series Data
- Smoothing: Using techniques like moving averages or exponential smoothing can help reduce the effect of noise and highlight underlying patterns.

- Outlier Detection: Identifying and removing outliers or extreme values caused by random events can help reduce noise and improve model accuracy.

- Noise Filtering:Techniques like Fourier transforms or wavelet transforms can be used to filter out high-frequency noise.

- Statistical Modeling:  Many forecasting models like ARIMA (AutoRegressive Integrated Moving Average) or Exponential Smoothing can account for noise by modeling it as random error or residuals.

# Stationarity in Time Series

Time series require the data to be stationary. Most of the models work on the assumption that time series is stationary.
Stationary has vary strict criteria
- constant mean, mean should be constant according to time
- constant variance, variance should be equal at time interval
- autocovariance that doesn't depend on time

*Stationarity* refers to the statistical property of a time series where the mean, variance, and covariance (or autocovariance) of the series do not change over time

*why it's important*
- Simplifies Modeling

#### Methods to check stationarity (ADF test, KPSS test).

Perform statistical tests, such as the Augmented Dickey-Fuller (ADF) test, Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test, or Phillips-Perron test, to formally test for stationarity. These tests help determine if a time series has a unit root, which indicates non-stationarity.The ADF test is a type of hypothesis test

Hypotheses
- Null Hypothesis (Hâ‚€): The time series has a unit root, i.e., the series is non-stationary.
- Alternative Hypothesis (Hâ‚): The time series does not have a unit root, i.e., the series is stationary.

- If the **p-value** is **less than 0.05**, we **reject the null hypothesis** and conclude that the series is stationary.
- If the **p-value** is **greater than 0.05**, we fail to reject the null hypothesis, meaning the series is likely non-stationary.

Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test

The **KPSS test** is another statistical test used to check for stationarity. Unlike the ADF test, the KPSS test has a **null hypothesis** that assumes the time series is stationary, and the **alternative hypothesis** assumes the series has a unit root (non-stationary).


#### Other Test for Stationary

- Rolling Statistics: Ploting moving average (At any instant t will take the average of time window eg last 12months)
- ADCF test: null hypothesis is that the time series is non-stationary. The test results comprise of a 'Test statistics' and some 'Critical values' if Test statistics< Critical values we reject the null hypothesis

ADF

```python
# Perform the Augmented Dickey-Fuller test
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print("ADF Test Results:")
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"\t{key}: {value}")
    
    if result[1] < 0.05:
        print("\nConclusion: The series is stationary (Reject H0).")
    else:
        print("\nConclusion: The series is non-stationary (Fail to reject H0).")

# Run ADF test on AirPassengers data
adf_test(airline['#Passengers'])
```

KPSS TEST

```python
# Perform the KPSS test
def kpss_test(series):
    result = kpss(series, regression='c', nlags="auto")
    print("\nKPSS Test Results:")
    print(f"KPSS Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print("Critical Values:")
    for key, value in result[3].items():
        print(f"\t{key}: {value}")
    
    if result[1] < 0.05:
        print("\nConclusion: The series is non-stationary (Reject H0).")
    else:
        print("\nConclusion: The series is stationary (Fail to reject H0).")

# Run KPSS test on AirPassengers data
kpss_test(airline['#Passengers'])
```

# Definition and explanation of the concepts of ACF and PACF

**Autocorrelation (ACF)**

Autocorrelation measures the linear relationship between a time series and its past values. In simpler terms, it tells you how much a data point at a specific time is related to data points at previous times.
- Lag: The time difference between the current data point and the past data point you're comparing it to. For example, a lag of 1 means you're comparing a data point to the one immediately before it.
- ACF: The Autocorrelation Function calculates the correlation coefficient between a time series and its lagged values at various lags. It helps you understand how the correlation changes as you consider data points further in the past.

**Partial Autocorrelation (PACF)**

Partial Autocorrelation measures the correlation between a time series and a specific lag, while controlling for the influence of intermediate lags. This helps isolate the direct relationship between the current data point and a specific point in the past, without the "noise" of other time points.
- PACF: The Partial Autocorrelation Function calculates these partial correlation coefficients for different lags.

# The purpose and methods of decomposition

Time series decomposition is a technique used to break down a time series into its constituent components. The primary purpose of decomposition is to understand the underlying patterns and drivers of the time series, which can be helpful for forecasting, analysis, and modeling.

Purpose of Decomposition:

1. **Identifying Components:**  Time series data often consists of a mix of different patterns:
   - Trend: The long-term direction of the data (upward, downward, or flat).
   - Seasonality: Regular, repeating patterns within the data (e.g., monthly sales spikes during holidays).
   - Cyclicity: Longer-term fluctuations that are not strictly periodic (e.g., business cycles).  These are harder to identify and separate from the trend.
   - Residual (or Noise): The random, irregular fluctuations that are left over after the other components have been removed.

2. **Simplifying Analysis:** By separating these components, it becomes easier to analyze each one individually. For example, you can study the trend to understand the overall growth or decline of a phenomenon, or you can analyze the seasonal component to predict recurring patterns.

3. **Improving Forecasting:** Many forecasting methods work best when the data is stationary (no trend or seasonality). Decomposition can be used as a preprocessing step to remove these components, making the data more suitable for forecasting. You can then forecast each component separately and combine the forecasts to produce an overall forecast.

4. **Understanding Drivers:** Decomposing a time series can help identify the factors that are driving the different components. For instance, the trend might be influenced by economic factors, while seasonality might be driven by weather patterns or consumer behavior.

Methods of Decomposition:

1. **Moving Average:** This is a simple method that smooths the data by calculating a moving average over a certain window. The moving average represents the trend component. The difference between the original data and the moving average is considered the seasonal and residual components. This method is less suitable for data with strong seasonality.

2. **Classical Decomposition:** This method assumes that the time series is composed of an additive or multiplicative combination of its components.  It first estimates the trend using moving averages, then removes the trend to estimate the seasonal component. The residual is what remains after removing the trend and seasonal components.  This method is sensitive to outliers and assumes a stable seasonal pattern.

3. **Seasonal Decomposition of Time Series (STL):** STL is a more robust method developed by Cleveland et al. It uses locally weighted scatterplot smoothing (LOWESS) to estimate the trend and seasonal components. STL can handle both additive and multiplicative seasonality, and it is less sensitive to outliers than classical decomposition. It also allows for the seasonal component to change over time.  This is generally the preferred method.

4. **X-11 and X-13ARIMA-SEATS:** These are statistical methods developed by the U.S. Census Bureau. They are widely used for seasonal adjustment of economic data. They are more complex than the other methods but offer more sophisticated handling of seasonality, trend, and calendar effects.



Discuss Classical and STL Decomposition

Classical Decomposition:

 **Concept**: Classical decomposition assumes that a time series can be expressed as either an additive or multiplicative combination of its components (trend, seasonality, and residual). It uses moving averages to estimate the trend and then extracts the seasonal component. The remainder is considered the residual

**Method**:
- Trend Estimation: A centered moving average is calculated to smooth out the data and estimate the trend. The window size of the moving average should match the periodicity of the seasonality (e.g., 12 for monthly data). Â  
- Seasonal Component Estimation: The seasonal component is calculated by subtracting the trend from the original data (additive) or dividing the original data by the trend (multiplicative). These seasonal values are then averaged for each period (e.g., all January values, all February values, etc.) to get a representative seasonal pattern. Â  
- Residual Calculation: The residual is the difference (additive) or ratio (multiplicative) between the original data and the combined trend and seasonal components

Seasonal Decomposition of Time Series (STL):

**Concept**: STL is a more sophisticated and robust method that uses locally weighted scatterplot smoothing (LOWESS) to decompose the time series.  It's more flexible and can handle more complex situations than classical decomposition.

**Method**:
- Trend Estimation: STL uses LOWESS to smooth the data and estimate the trend. LOWESS is a non-parametric regression technique that fits a local regression at each point in the data. Â  
- Seasonal Component Estimation: STL iteratively estimates the seasonal component. It removes the trend from the data and then uses LOWESS smoothing to estimate the seasonal component. This process is repeated until convergence. Â  
- Residual Calculation: The residual is calculated as the difference (additive) or ratio (multiplicative) between the original data and the combined trend and seasonal components

Which method to use

**STL**:  Generally preferred due to its robustness, flexibility, and ability to handle evolving seasonality.  It's a good default choice for most time series decomposition tasks

**Classical** Decomposition: Only consider if you have a very simple time series with a clear additive or multiplicative structure, constant seasonality, and you are not concerned about outliers.  It can be useful for quick, preliminary analysis
